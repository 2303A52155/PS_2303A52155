HISTOGRAM

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/car_data_250.csv")

bins = 10

plt.figure(figsize=(12, 6))
plt.hist(df['Price'], bins=bins, color='skyblue', edgecolor='black', alpha=0.75)
plt.xlabel('Price (USD)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Car Prices', fontsize=14)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

price_bins = pd.cut(df['Price'], bins=bins)
most_common_range = price_bins.value_counts().idxmax()
print(f"Most common price range: {most_common_range}")

```

SCATTER PLOT

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/car_data_250.csv')

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Horsepower', y='Mileage', hue='Fuel Type', palette='deep')
sns.regplot(data=df, x='Horsepower', y='Mileage', scatter=False, color='black', line_kws={"color": "red"})
plt.title('Horsepower vs Mileage')
plt.xlabel('Horsepower')
plt.ylabel('Mileage')
plt.legend(title='Fuel Type')
plt.show()
```

BAR CHART VISUALIZATION

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/car_data_250.csv')

brand_counts = df['Brand'].value_counts()

most_common_brand = brand_counts.idxmax()
highest_count = brand_counts.max()

plt.figure(figsize=(12, 6))
brand_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Cars per Brand')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

print(f'Highest: {most_common_brand} ({highest_count} cars)')
```

GROUPED BAR CHART

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/car_data_250.csv')

fuel_year_counts = df.groupby(['Year', 'Fuel Type']).size().unstack()

fuel_year_counts.plot(kind='bar', figsize=(12, 6), colormap='viridis', edgecolor='black')
plt.title('Number of Cars by Fuel Type and Manufacturing Year')
plt.xlabel('Manufacturing Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Fuel Type')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
